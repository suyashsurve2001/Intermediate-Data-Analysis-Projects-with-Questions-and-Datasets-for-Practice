2.FOr speeding , were men or women stopped more often

```python
df[df.violation=="Speeding"].driver_gender.value_counts()
```

Out of the total stops for speeding, men were stopped 25,517 times, while women were stopped 11,686 times. Therefore, it can be inferred that males tend to be stopped for speeding more frequently.

3.Does gender affect who gets searched during a stop

```python
df['search_conducted'].value_counts()
```

```python
df.groupby("driver_gender").search_conducted.sum()
```

The likelihood of being searched during a stop is influenced by gender, as evidenced by the fact that out of 63,056 stops, 366 involved females, and 2,113 involved males.

4.What is the mean  of stop duration(Mapping + data type casting)

```python
df.stop_duration.value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5 ,'16-30 Min' : 24 , '30+ Min'  : 45})
     
```

```python
df['stop_duration'].mean()
```

The average duration of a stop resulting in a search is 12.5 minutes.
